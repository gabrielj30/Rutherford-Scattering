# Sample Data Plot

```python
# a is theta_0 = theta_1 + theta_2

y = np.linspace(1,20,20)
theta_1 = 19.1
theta_2 = np.arctan(2.5/y)

a = theta_1 + theta_2

def f_y(theta_2, a):
  return (np.cos(theta_2) * (np.sin(theta_2))**2) / ((np.sin(a/2))**4)

print(theta_2)

plt.xlabel("Measurements in cm")
plt.ylabel('F(y)')
plt.title('F(y) vs Measure (cm) [SAMPLE DATA]')
plt.plot(y, f_y(theta_2, a), color = "orange")
plt.show()
```
